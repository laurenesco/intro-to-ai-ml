```python
# Import Linear Discriminant Analysis for dimensionality reduction and datasets
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.datasets import load_wine
# Import pyplot for plotting graphs
import matplotlib.pyplot as plt

# Load the wine dataset
data = load_wine()
# Separate the features (X) and the target variable (y)
X, y = data.data, data.target
# Get the names of the different types of wine classes
target_names = data.target_names

# Initialize the LDA model
lda = LinearDiscriminantAnalysis()
# Fit the LDA model to the data and transform the features to a lower dimension
X_r2 = lda.fit(X, y).transform(X)

# Define a list of colors to use for the different classes
colors = ['navy', 'turquoise', 'darkorange']
# Create a new figure for plotting
plt.figure()

# Loop through each class and its corresponding color and label for plotting
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    # Scatter plot of the reduced data points, colored by class
    # 'y == i' selects data points belonging to class i
    plt.scatter(X_r2[y == i, 0], X_r2[y == i, 1], alpha=.8, color=color, label=target_name)

# Add a legend in the best location, with no shadow and 1 point per legend item
plt.legend(loc='best', shadow=False, scatterpoints=1)

# Add a title to the scatter plot
plt.title('LDA of Wine Recognition dataset')

# Display the plot
plt.show()
```
