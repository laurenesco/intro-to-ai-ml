```python
# Import PCA for dimensionality reduction and pyplot for plotting graphs
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Import StandardScaler for feature scaling
from sklearn.preprocessing import StandardScaler

# Load the wine dataset from scikit-learn datasets
from sklearn.datasets import load_wine
data = load_wine()

# Separate the features and target from the dataset
X_, y = data.data, data.target

# Get the names of the target classes
target_names = data.target_names

# Print the shape of the original feature matrix to show it has 13 features
print(X_.shape)  # Outputs: (number_of_samples, 13)
```

```python
# Standardize features by removing the mean and scaling to unit variance
# This step is necessary to perform PCA correctly
X = StandardScaler().fit_transform(X_)

# Initialize PCA, setting the number of components to 2 for 2D projection
pca = PCA(n_components=2)

# Apply PCA to the scaled data and project it onto the 2 principal components
X_r = pca.fit_transform(X)
```

```python
# Create a new figure for plotting
plt.figure()

# Define a list of colors for the scatter plot
colors = ['navy', 'turquoise', 'darkorange']
# Line width for the scatter points
lw = 2

# Plot the projection of the data in 2D, colored by the target class
# Iterate over each class to plot them on the graph
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
   plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=lw, label=target_name)

# Add a legend to the plot with the best location, no shadow, and 1 scatterpoint in the legend
plt.legend(loc='best', shadow=False, scatterpoints=1)

# Set the title of the plot
plt.title('PCA of Wine recognition dataset')

# Print the explained variance ratio of the first two principal components
print('explained variance ratio (first two components): %s' % str(pca.explained_variance_ratio_))

# Show the plot
plt.show()
```
