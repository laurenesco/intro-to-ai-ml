# Linear Regression Analysis with Python

```python
# Import the necessary libraries for data manipulation, statistical modeling, and plotting
import numpy as np  # For numerical operations
import pandas as pd  # For data manipulation
import statsmodels.api as sm  # For statistical models, specifically Ordinary Least Squares (OLS) Linear Regression
import matplotlib.pyplot as plt  # For plotting
```

```python
# Load the dataset from 'regression.csv'
df = pd.read_csv('regression.csv')
```

```python
# Define the predictor variable 'X' and the response variable 'y'
X = df['x']  # Predictor variable
y = df['y']  # Response variable

# Add a constant to the predictor variable to include an intercept in the model
X = sm.add_constant(X)

# Fit the linear regression model using the OLS method
lr_model = sm.OLS(y, X).fit()

# Plotting the regression results
plt.xlabel('x')  # Label for the x-axis
plt.ylabel('y')  # Label for the y-axis
plt.title('Linear Regression (OLS)')  # Title for the plot

# Define lists of x and y axis ticks for better visualization
xt = np.arange(0, df['x'].max() + 5, 5)  # X-axis ticks based on the range of 'x'
yt = np.arange(0, df['y'].max() + 10, 10)  # Y-axis ticks based on the range of 'y'

# Set ticks and axis limits for the plot to enhance readability
plt.xticks(xt)
plt.yticks(yt)
plt.xlim(0, df['x'].max() + 5)  # X-axis limits
plt.ylim(0, df['y'].max() + 10)  # Y-axis limits

# Plot the data points
plt.scatter(df['x'], df['y'], color='blue')  # Data points in blue

# Predict and plot the regression line
y_pred = lr_model.predict(X)
plt.plot(df['x'], y_pred, color='red')  # Regression line in red

plt.show()  # Display the plot
```
