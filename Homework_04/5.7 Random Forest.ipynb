```python
# This exercise is based on the Abalone dataset, which can be downloaded from OA 5.12.  
#
# The task is to predict the age of abalone from physical measurements. 
# The age of abalone  is determined by cutting the shell through the cone, 
# staining it, and counting the number of rings through a microscope â€“ a 
# boring and time-consuming task. Other measurements,  which are easier to 
# obtain, are used to predict the age. Table 5.7 shows the list of attributes
# that are available in the current dataset.  The original data examples had 
# a couple of missing values, which were removed (the  majority having the 
# predicted value missing), and the ranges of the continuous values have  
# been scaled for use with an ANN (by dividing by 200).  
#
# Use this dataset to predict the age of abalone from the given attributes. 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

# Load, evaluate and prepare the data
colnames = ['Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Schucked weight', 'Viscera weight', 'Shell weight', 'Rings']
df = pd.read_csv("../Datasets/Code_and_Data_(Chapter_05)/Code and Data (Chapter 05)/abalone.csv", names=colnames)
# print (df)

X = df.drop(columns='Sex')
y = df.Sex
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
# Visualize the data
plt.bar(['M', 'F', 'I'], [len(y[y == 'M']), len(y[y == 'F']), len(y[y == 'I'])], tick_label=['Male', 'Female', 'Infant'], color=['#9E768F', '#9FA4C4', '#B3CDD1'])
plt.grid(axis='y', linestyle='--', alpha=0.6)
```

```python
# Let us create the Random Forest model, train it, and make some predictions
rfc = RandomForestClassifier(n_estimators=500)
rfc.fit(X_train, y_train)
predictions = rfc.predict(X_test)

# Evaluate the model
print(accuracy_score(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```
